```python
# construting a pivot table
pivotDF = DataFrame({"Commodity":['A','A','A','A','B','B','B','C','C'],
                     "Amount":[10,15,5,20,10,10,5,20,30],
                     "Partner":[10,15,5,20,10,10,5,20,30],
                     "Flow":["X","Y","X","Y","X","Y","X","X","Y"]
}
)
pivotDF
```

```python
pivot_table(pivotDF,
                   index =['Commodity','Partner'],
                   values='Amount',
                    aggfunc=sum
           )
```

```python
#pivot table with row and column
pivot_table(pivotDF,
           index =['Commodity','Partner'],
           columns='Flow',
           values ='Amount',
           aggfunc=sum)
```

```python
# margins=True. adds row and column
pivot_table(pivotDF,
            index = ['Commodity','Partner'],
             columns ='Flow',
             values ='Amount',
             aggfunc=sum,
             margins =True
            )


```
