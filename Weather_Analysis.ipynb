Weather Underground keeps historical weather data collected in many airports around the world. 

http://www.wunderground.com/history


```python
import warnings
warnings.simplefilter('ignore', FutureWarning)

from pandas import *
# reading the data.
BeijingWeather = read_csv('Beijing_PEK_2014.csv',skipinitialspace=True)
BrasiliaWeather = read_csv('Brasilia_BSB_2014.csv',skipinitialspace=True)
CapeWeather = read_csv('CapeTown_CPT_2014.csv',skipinitialspace=True)
DelhiWeather = read_csv('Delhi_DEL_2014.csv',skipinitialspace=True)
MoscowWeather = read_csv('Moscow_SVO_2014.csv',skipinitialspace=True)

# editing columns
# remove break tags from the column names.
BeijingWeather = BeijingWeather.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
BrasiliaWeather = BrasiliaWeather.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
CapeWeather = CapeWeather.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
DelhiWeather = DelhiWeather.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
MoscowWeather = MoscowWeather.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})

# remove the break tags from the column values.
BeijingWeather['WindDirDegrees'] = BeijingWeather['WindDirDegrees'].str.rstrip('<br />')
BrasiliaWeather['WindDirDegrees'] = BrasiliaWeather['WindDirDegrees'].str.rstrip('<br />')
CapeWeather['WindDirDegrees'] = CapeWeather['WindDirDegrees'].str.rstrip('<br />')
DelhiWeather['WindDirDegrees'] = DelhiWeather['WindDirDegrees'].str.rstrip('<br />')
MoscowWeather['WindDirDegrees'] = MoscowWeather['WindDirDegrees'].str.rstrip('<br />')


# change the values of WindDirDegrees to int64
BeijingWeather['WindDirDegrees'] = BeijingWeather['WindDirDegrees'].astype('float64')   
BrasiliaWeather['WindDirDegrees'] = BrasiliaWeather['WindDirDegrees'].astype('float64')   
CapeWeather['WindDirDegrees'] = CapeWeather['WindDirDegrees'].astype('float64')   
DelhiWeather['WindDirDegrees'] = DelhiWeather['WindDirDegrees'].astype('float64')   
MoscowWeather['WindDirDegrees'] = MoscowWeather['WindDirDegrees'].astype('float64')   


# converting the 'date' column from object data type to date_time data type
BeijingWeather['Date'] =to_datetime(BeijingWeather['Date'])
BrasiliaWeather['Date'] =to_datetime(BrasiliaWeather['Date'])
CapeWeather['Date'] =to_datetime(CapeWeather['Date'])
DelhiWeather['Date'] =to_datetime(DelhiWeather['Date'])
MoscowWeather['Date'] =to_datetime(MoscowWeather['Date'])

# setting the date column as the index of the tables.
BeijingWeather.index =BeijingWeather['Date']
BrasiliaWeather.index =BrasiliaWeather['Date']
CapeWeather.index =CapeWeather['Date']
DelhiWeather.index =DelhiWeather['Date']
MoscowWeather.index =MoscowWeather['Date']


BrasiliaWeather['Date'].head(10)
```
