```python
#  if else statements in python
def expandCountry(name):
    if name =='UK':
        return 'United Kingdom'
    elif name =='USA':
        return 'United States'
    elif name=='St.Lucia':
        return 'Saint Lucia'
    else:
        return name
# test cases
expandCountry('St.Lucia')=='Saint Lucia'
```

```python
# test cases.
expandCountry('UK')=='United kingdom'
expandCountry('UK')=='UnitedKingdom'
expandCountry('UK')=='  United Kingdom'
expandCountry('UK')=='United Kingdom'
```

```python
# if loop without elif and else
def CountryName(name):
    if name=='UK':
       name =='United Kingdom'
    if name =='USA':
       name == 'United States'
    return name
CountryName('India')=='India'
```
