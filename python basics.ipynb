```python
# recursions
def tri_recursion(k):
    if (k>0):
        print("first ",k)
        print("tri_recursion",tri_recursion(k-1))
        result= k+tri_recursion(k-1)
        print("result",result)
        print("tri_recursion(k-1)",tri_recursion(k-1))

    else:
        result =0
    return result
tri_recursion(6)
```
